```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configurao visual
sns.set_theme(style="darkgrid")

# Carregar base
df = pd.read_csv('Superstore.csv')

# Ver primeiras linhas
display(df.head())
df.info()

plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Segment', palette='Set2')
plt.title('Quantidade de clientes por Segmento')
plt.xlabel('Segmento')
plt.ylabel('Quantidade')
plt.show()


plt.figure(figsize=(14,10))
sns.countplot(data=df, x='State', palette='viridis')
plt.title('Quantidade de pedidos por Estado')
plt.xlabel('Estados')
plt.ylabel('Quantidade')
plt.xticks(rotation=90)
plt.show()

plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Category', palette='crest')
plt.title('Quantidade de Registros por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Quantidade')
plt.show()

plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Category', y='Profit', estimator='sum', palette='viridis')
plt.title('Lucro Total por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Lucro Total')
plt.show()


top_profit = df.groupby('Sub-Category')['Profit'].sum().sort_values()

plt.figure(figsize=(12,8))
top_profit.plot(kind='barh')
plt.title('Lucro por Sub-Categoria')
plt.xlabel('Lucro Total')
plt.ylabel('Sub-Categoria')
plt.show()

```
